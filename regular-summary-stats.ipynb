# Now lets add distinct_percentage
we just want to divide distinct_count by length
but we have no decent way to insert that calculation, we have to rewrite the entire function

```python
def col_stats(ser):
    
    l = len(ser)
    values = ser.value_counts()
    distinct_count = values[values == 1].sum()

    base_stats = dict(
        nan_count = ser.isna().sum(),
        distinct_count = distinct_count,
        distinct_percentage = distinct_count / l,
        dtype = ser.dtype)
    if pd.api.types.is_numeric_dtype(ser):
        base_stats.update(dict(
                min = ser.min(),
                mean = ser.mean(),
                max = ser.max(),
                mode = ser.mode()[0]
        ))
    return base_stats

def summary(df, size=50):
    #we'll start with the first 50 rows as a smoke test
    if size > 0:
        small_df = df.iloc[:size]
    else:
        #use the whole thing
        small_df = df
    return pd.DataFrame(dict([[col, col_stats(small_df[col])] for col in small_df.columns]))

summary(df)
```

# What about individual stats functions
summary is modified to recieve a list of functions to apply in order

```python

def base_col_stats(ser):
    
    l = len(ser)
    values = ser.value_counts()
    distinct_count = values[values == 1].sum()

    base_stats = dict(
        nan_count = ser.isna().sum(),
        distinct_count = distinct_count,
        dtype = ser.dtype)
    if pd.api.types.is_numeric_dtype(ser):
        base_stats.update(dict(
                min = ser.min(),
                mean = ser.mean(),
                max = ser.max(),
                mode = ser.mode()[0]
        ))
    return base_stats

def distinct_percentage(ser):
    
    l = len(ser)
    values = ser.value_counts()
    distinct_count = values[values == 1].sum()

    return dict(
        distinct_percentage = distinct_count / l)
    return base_stats

def summary_funcs(df, funcs, size=50):
    if size > 0:
        small_df = df.iloc[:size]
    else:
        #use the whole thing
        small_df = df
    
    facts_dict = {}
    for col in small_df.columns:
        facts = {}
        for func in funcs:
            facts.update(func(small_df[col]))
        facts_dict[col] = facts
    return pd.DataFrame(facts_dict)

summary_funcs(df, [base_col_stats, distinct_percentage])
```

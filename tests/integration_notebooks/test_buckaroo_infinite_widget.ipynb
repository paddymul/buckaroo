```python
import buckaroo
import pandas as pd
from buckaroo import BuckarooInfiniteWidget

# Create test data
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35],
    'score': [85.5, 92.0, 78.3]
})
print(f"✅ Created DataFrame with shape: {df.shape}")

# Display the widget
widget = BuckarooInfiniteWidget(df)
print("✅ BuckarooInfiniteWidget created successfully")
widget
```

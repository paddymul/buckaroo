```python
import buckaroo
import polars as pl
from buckaroo.polars_buckaroo import PolarsBuckarooInfiniteWidget

# Create a large predictable DataFrame for testing infinite scroll
# Row i has: int_col = i + 10, str_col = f"foo_{i + 10}"
# 2000 rows ensures we need multiple data fetches to satisfy scroll requests
N_ROWS = 2000
df = pl.DataFrame({
    'row_num': list(range(N_ROWS)),
    'int_col': [i + 10 for i in range(N_ROWS)],
    'str_col': [f'foo_{i + 10}' for i in range(N_ROWS)]
})
print(f"✅ Created DataFrame with shape: {df.shape}")
print(f"   First row: row_num=0, int_col=10, str_col='foo_10'")
print(f"   Last row: row_num={N_ROWS-1}, int_col={N_ROWS-1+10}, str_col='foo_{N_ROWS-1+10}'")

# Display the widget with transcript recording enabled
widget = PolarsBuckarooInfiniteWidget(df, record_transcript=True)
print("✅ PolarsBuckarooInfiniteWidget created successfully with transcript recording enabled")
widget
```
